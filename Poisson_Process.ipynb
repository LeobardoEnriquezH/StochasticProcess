```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Process intensity
lamb=7
#Events
n=50
#UNIFORM time bwtween events
TUnif=np.random.random(n)
TExp=-(1/lamb)*np.log(TUnif).cumsum()
```

```python
TExp=np.insert(TExp,0,0)
TExp
```

```python
#Las ocurrencias:
Occ=np.ones(n).cumsum()
Occ=np.insert(Occ,0,0)
plt.figure(figsize=[20,10])
plt.step(TExp, Occ, linestyle='--', marker='.', color='olive')
plt.title('Proceso Poisson Homogneo')
plt.xlabel('Tiempo')
plt.ylabel('Nmero de ocurrencias')
plt.grid('on')
plt.show()
```

```python
#Ocurrencias: suma de N(0,1) y Unif(0,1)
Occ=np.random.normal(0,1,n).cumsum()
Occ=np.insert(Occ,0,0)
plt.figure(figsize=[20,10])
plt.step(TExp, Occ, linestyle='--', marker='.', color='olive')
plt.title('Proceso Poisson Compuesto con N(0,1)')
plt.xlabel('Tiempo')
plt.ylabel('Nmero de ocurrencias')
plt.grid('on')
plt.show()
```

```python
#Proceso Pisson: Proceso de Cramer - Lundberg

#Intensidad del proceso subyacente
lamb=1
#Numero de eventos:
n=20
#Los tiempos uniformes entre eventos:
TUnif=np.random.random(n)
TExp=-(1/lamb)*np.log(TUnif).cumsum()
TExp=np.insert(TExp,0,0)

Occ=np.random.exponential(1,n).cumsum()
Occ=np.insert(Occ,0,0)
x=10
xs=x*np.ones(n+1)
c=1
plt.figure(figsize=[20,10])
plt.plot(TExp, xs+c*TExp-Occ, linestyle='--', marker='.', color='olive')
plt.title('Proceso de Cramer-Lundberg con exp(1)')
plt.xlabel('Tiempo')
plt.ylabel('Nmero de ocurrencias')
plt.grid('on')
plt.show()
```
