```python
import numpy as np
import random as rm
```

```python
p = 0.5
A= np.array([[1-p,p,0,0],[0,0,1-p,p],[1-p,p,0,0],[0,1,0,0]])
```

```python
# Clasificamos los estados
state = [0,1,2,3]
```

```python
# Simulacion de la cadena
#Definimos el numero de simulaciones
n= 1000
#Definimos el nuestro estado inicial
start_state= 0
print(state[start_state], "--->", end = " ")
prev_state = start_state
walk = []
while n-1:
    curr_state = np.random.choice([0,1,2,3],p = A[prev_state])
    walk.append(state[curr_state])
    print(state[curr_state], "--->", end = " ")
    prev_state = curr_state
    n-=1
print("stop")
walk.insert(0,start_state)
print(walk)
```

```python
# Calculamos Nj*M para cada estado
N = []
for i in range(len(state)):
    N.append(walk.count(i))
print(N)    
```

```python
C = []
M = 1000
for i in range(len(N)):
    C.append(N[i]/M)
print(C)
```
